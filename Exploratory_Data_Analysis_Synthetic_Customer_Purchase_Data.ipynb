# Exploratory Data Analysis of Synthetic Customer Purchase Data

This notebook demonstrates:
- Synthetic data generation using NumPy and Pandas
- Data inspection and descriptive statistics
- Exploratory Data Analysis (EDA)
- Visualization using Matplotlib and Seaborn


```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Reproducibility
np.random.seed(42)
```

## Step 1: Generate Synthetic Dataset

```python
num_records = 500

data = {
    "CustomerID": np.arange(1, num_records + 1),
    "Age": np.random.randint(18, 70, num_records),
    "PurchaseAmount": np.round(np.random.uniform(10, 500, num_records), 2),
    "ProductCategory": np.random.choice(["Electronics", "Clothing", "Home", "Sports"], num_records),
    "CustomerSegment": np.random.choice(["Regular", "Premium", "VIP"], num_records)
}

df = pd.DataFrame(data)
df.head()
```

## Step 2: Initial Data Inspection

```python
print("Data Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())
print("\nDescriptive Statistics:\n", df.describe())
```

## Step 3: Statistical Analysis

```python
mean_purchase = df['PurchaseAmount'].mean()
median_purchase = df['PurchaseAmount'].median()
mode_purchase = df['PurchaseAmount'].mode()[0]

print("Mean Purchase Amount:", mean_purchase)
print("Median Purchase Amount:", median_purchase)
print("Mode Purchase Amount:", mode_purchase)

category_freq = df['ProductCategory'].value_counts()
print("\nProduct Category Frequency:\n", category_freq)
```

## Step 4: Visualizations

```python
sns.set()

# Histogram of Age
plt.figure()
plt.hist(df['Age'], bins=10)
plt.title('Distribution of Customer Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Bar chart for ProductCategory counts
plt.figure()
df['ProductCategory'].value_counts().plot(kind='bar')
plt.title('Count of Records per Product Category')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.show()

# Box plot of PurchaseAmount across CustomerSegment
plt.figure()
sns.boxplot(x='CustomerSegment', y='PurchaseAmount', data=df)
plt.title('Purchase Amount by Customer Segment')
plt.show()
```

## Step 5: EDA Summary

The exploratory data analysis of the synthetic customer purchase dataset reveals several key trends. The age distribution appears balanced across different ranges, suggesting diversity among customers. Descriptive statistics indicate a wide spread in purchase amounts, highlighting varied spending behavior. The mean and median purchase values are relatively close, implying a roughly symmetric distribution. Product category counts show differences in purchasing frequency across categories. The box plot suggests that Premium and VIP segments may demonstrate higher variability in spending compared to Regular customers. Overall, this EDA highlights how basic statistical analysis and visualization help uncover trends and prepare data for further analysis.
